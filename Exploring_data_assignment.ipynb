### visualize the relation between crime rates and a selection of features

```python
selection = ['Crimes_next_year', 'Schools_within_3k', '45-65_yo', 'Household_size', 'Households_one_person', 'Households_below_poverty_level', '0-15_yo']
df_train.loc[:,selection].isnull().sum()
```

```python
_impute_value = df_train.loc[:,selection].median()
df_temp = df_train.loc[:,selection].fillna(value = _impute_value)
df_temp.loc[:,selection].isnull().sum()
```

```python
pd.plotting.scatter_matrix(df_temp.astype('int64'), figsize=(12,12), alpha=0.2);
```
